```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

boston_df = pd.read_csv('/content/boston.csv')
boston_df.head()
#boston_df.isnull().sum()
#boston_df.info()

corr_matrix = boston_df.corr().round(2)
corr_matrix

x1 = boston_df['LSTAT']
x2 = boston_df['RM']
y = boston_df['MEDV']

plt.figure(figsize = (10,6))
plt.scatter(x1, y)

plt.xlabel('Процент населения с низким социальным статусом', fontsize = 15)
plt.ylabel('Медианная цена недвижимости, тыс. долларов', fontsize = 15)
plt.title('Социальный статус населения и цены на жилье', fontsize = 18)

plt.figure(figsize = (10,6))
plt.scatter(x2, y)

plt.xlabel('Среднее количество комнат', fontsize = 15)
plt.ylabel('Медианная цена недвижимости, тыс. долларов', fontsize = 15)
plt.title('Среднее количество комнат и цены на жилье', fontsize = 18)

X = boston_df[['RM', 'LSTAT', 'PTRATIO', 'TAX', 'INDUS']]
y = boston_df['MEDV']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
random_state = 42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(y_pred[:5])

print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
print('R2:', np.round(metrics.r2_score(y_test, y_pred), 2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

credit_df = pd.read_csv('/content/credit_risk_dataset.csv')
credit_df.head()
#credit_df.isnull().sum()
#credit_df.info()

#credit_df.isnull().sum()

credit_df['person_emp_length'] = credit_df['person_emp_length'].fillna(0)
credit_df['person_emp_length'] = credit_df['person_emp_length'].dropna()

credit_df['loan_int_rate'] = credit_df['loan_int_rate'].fillna(0)
credit_df['loan_int_rate'] = credit_df['loan_int_rate'].dropna()

#corr_matrix = credit_df.corr().round(2)
#corr_matrix
#person_age-cb_person_cred_hist_length (0.86), loan_percent_income-loan_amnt (0.57), loan_status-loan_int_rate(0.34)

x1 = credit_df['person_age']
y = credit_df['cb_person_cred_hist_length']

plt.figure(figsize = (10,6))
plt.scatter(x1, y)

plt.xlabel('Возвраст', fontsize = 15)
plt.ylabel('Кредитная история', fontsize = 15)
plt.title('Возраст человека и его кредитная история', fontsize = 18)

X = credit_df[['cb_person_cred_hist_length', 'person_income', 'person_emp_length']]
y = credit_df['person_age']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
random_state = 42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred[:5])

print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred))) 
print('R2:', np.round(metrics.r2_score(y_test, y_pred), 2))
```
